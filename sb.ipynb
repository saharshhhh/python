```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="teal")
plt.title("Distribution of total bills")
```

KDE PLOT: KERNEL DENSITY ESTIMATION

```python
sns.kdeplot(data=tips, x='total_bill', fill=True, color='teal')
```

BAR PLOT

```python
sns.barplot(data=tips,x='day', y='total_bill',legend= False,hue="sex" ,palette="viridis")
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", palette="coolwarm")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue='day', style="time", palette='deep')
```

```python
sns.lineplot(data=tips, x='size', y='tip', marker='*',color='green')
# plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
```

```python
numtips=tips.select_dtypes(["number"])
correlation_matrix=numtips.corr()
sns.heatmap(correlation_matrix, annot=True,cmap="coolwarm",fmt=".2f")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.DataFrame(f)
subset=df[['Age','Fare','Pclass','Survived']]
sns.pairplot(data=subset, hue='Survived', diag_kind='kde')
plt.suptitle("Pairwise Relationships", y=1.02)
```

```python
numeric_cols= df.select_dtypes(include='number')
corr_matrix=numeric_cols.corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
```
