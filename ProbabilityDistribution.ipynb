```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
p=0.5
data=bernoulli.rvs(p,size=1000)

plt.bar([0,1], [sum(data==0), sum(data==1)],color=['red','green'], tick_label=["Failure(0)","Success(1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.stats import norm
myu=70
sigma=10
x=np.linspace(myu - 4*sigma, myu + 4*sigma, 100)
pdf= norm.pdf(x,myu, sigma)
plt.plot(x,pdf,color='orange',label=f"N({myu},{sigma}**2)")
plt.legend()
plt.grid()
# plt.show()
```

```python
z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='black', label="Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0, color='red', linestyle='-.', label="Mean (z=0)")
plt.grid()
plt.legend()
plt.show()
```

CENTRAL LIMIT THEOREM

```python
import seaborn as sns
np.random.seed(42)
income_data=np.random.exponential(scale=50, size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data, bins=50,kde=True, color="black")
sample_means=[]
n_samples=1000
sample_size=30

for _ in range(n_samples):
    sample= np.random.choice(income_data,size=sample_size)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(8,4))
sns.histplot(sample_means,bins=50, kde=True, color='teal')
```
